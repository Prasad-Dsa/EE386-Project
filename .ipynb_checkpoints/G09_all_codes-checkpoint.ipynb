```python
#all import statements
import numpy as np
import IPython

from scipy import signal as sp
from scipy.io import wavfile
from scipy.fft import fft, ifft

from matplotlib import pyplot as plt
from matplotlib import rcParams
%matplotlib inline
plt.style.use(['seaborn-deep'])
plt.rcParams.update({
   # "text.usetex": True,
    "font.serif": ["cm"],
    "mathtext.fontset": "cm",
    "figure.figsize": (12,12),
    "font.size": 24})

import sounddevice as sd
import pywt


#User-defined functions
def mag_spectrum_plot(x, Fs, max_freq=None, label1=None, title1=None, save=None, color=None):
    C = int(len(x) / 2)
    if max_freq:
        C = int(C * max_freq / float(Fs) * 2)
    X = np.abs(fft(x)[0:C])
    N = (np.arange(0, C)/ len(x))*Fs
    plt.plot(N, X/len(x), color=color, label=label1)#normalization of magnitude
    plt.title(title1)
    plt.xlabel(r'$F(Hz)$')
    plt.ylabel(r'$|X(F)|$')
    if save:
        plt.savefig(save+'.png', format='png')
    plt.tight_layout()
    
def disc_spectrum_plot(x, label=None, title=None):
    N=len(x)
    f=np.arange(0, N//2)/N
    X=np.abs(fft(x))[0:len(f)]
    plt.plot(f, 20*np.log10(X), label=label)
    plt.title(title)
    plt.xlabel(r'Frequency $f$ (cycles/sample)')
    plt.xlim((0, 0.5))
    plt.tight_layout()
```

```python
#DEMO 1: different musical signals with similar Fourier representation
freqs=[440, 600, 880]
sig1=[]
sig2=[]
t=np.arange(0, 1, 0.0001)
for i in range(len(freqs)):
    sig1=np.hstack((sig1, (np.cos(2*np.pi*freqs[i]*t))))
    sig2=np.hstack((sig2, (np.cos(2*np.pi*freqs[len(freqs)-1-i]*t))))
wavfile.write('signal1.wav', 10000, (32767*np.array(sig1)).astype(np.int16) )
wavfile.write('signal2.wav', 10000, (32767*np.array(sig2)).astype(np.int16) )
plt.subplot(211)
mag_spectrum_plot(sig1, 10000, max_freq=2000, title1=r'Magnitude Spectrum Plot for Signal 1')
plt.subplot(212)
mag_spectrum_plot(sig2, 10000, max_freq=2000, title1=r'Magnitude Spectrum Plot for Signal 2')
plt.savefig('Demo1.png')
```

```python
#spectrogram code in function form: pass argument 0 for rectangular window, 1 for Hamming, 2 for Hanning and 3 for Gaussian
#Default window: Rectangular
#Default frame length: 1000
#Default overlap: 128\
#Default frequency range: 0 to 2kHz

def custom_spect(wavfile_name, window=None, std_for_gaussian=None, window_length=None, overlap_number=None, title=None, save=None, freq_range=None):
    Fs, x=wavfile.read(wavfile_name+'.wav')
    L=1000
    if window_length:
        L=window_length
    noverlap=128
    if overlap_number:
        noverlap=overlap_number
        x=x/32767
        n=len(x)
    w=sp.windows.boxcar(L)
    std=40
    if std_for_gaussian:
        std=std_for_gaussian
    windows=[sp.windows.boxcar(L), sp.windows.hamming(L), sp.windows.hann(L), sp.windows.gaussian(L, std)]
    if window:
        w=windows[window]
    j=0
    while True:
        if j==0:
            end=L
            prev_start=0
        else:
            end=prev_start+2*L-noverlap
            prev_start=prev_start+L-noverlap
        j=j+1
        if end>=n:
            break
    N=j
    freqs=np.linspace(0, int(Fs), L)
    X_fin=np.zeros((L, N))
    print(X_fin.shape)
    for i in range(N):
        if i==0:
            end=L
            prev_start=0
        else:
            end=prev_start+2*L-noverlap
            prev_start=prev_start+L-noverlap
        if end>=n:
            break
        else:
            Xtemp=(np.power(np.abs(fft(w*x[prev_start:end])), 2))
            X_fin[:, i]=Xtemp
#PURE PHASE TERM MULTIPLYING THE EXPRESSION IS IGNORED BECAUSE ONLY MAGNITUDE IS DESIRED
    plt.rcParams['figure.figsize'] = [15, 12]
    plt.pcolormesh(np.arange(N)*(L-noverlap)/(Fs), freqs, (X_fin), shading='gouraud', edgecolors=None, cmap=None)
    f_lims=(0, 2000)
    if freq_range:
        f_lims=freq_range
    plt.ylim(f_lims)
    plt.title(title)
    plt.xlabel(r'time $t$ in seconds')
    plt.ylabel(r'$F(Hz)$')
    if save:
        plt.savefig(save+'.png')
    plt.show()
```

```python
#Wavelet Generating Function
def mother(w, eta, t, tau, s):
    #t=np.linspace(0, to, int(Fs*to), endpoint=False)
    t1=(t-tau)/s
    psi=(1/np.sqrt(s))*(1/w)*np.exp(-np.pi*((t1/w)**2))*np.exp(1j*2*np.pi*eta*(t1/w))
    return psi

t=np.linspace(0, 5, 1000*5, endpoint=False)
psi=mother(0.25, 2, t, 2, 4)
plt.plot(t, np.real(psi))
#plt.plot(t, np.imag(psi))
print(len(psi))
print(np.sum(psi*np.conj(psi)))
plt.title('Child Wavelet with $b=2s$ and $a = 4$')
plt.xlabel('time (s)')
plt.ylabel(r'$\psi _{a, b}(t)$')
plt.savefig('child_42.png')
```

```python


from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng()
```

```python
#Generate a test signal, a 2 Vrms sine wave at 50Hz corrupted by 0.001 V**2/Hz of white noise sampled at 1024 Hz.
fs = 1024

N = 10*fs

nperseg = 512

amp = 2 * np.sqrt(2)

noise_power = 0.001 * fs / 2

time = np.arange(N) / float(fs)

carrier = amp * np.sin(2*np.pi*50*time)

noise = rng.normal(scale=np.sqrt(noise_power),

                   size=time.shape)

x = carrier + noise
```

```python
x = carrier + noise
#Compute the STFT, and plot its magnitude
f, t, Zxx = signal.stft(x, fs=fs, nperseg=nperseg)

plt.figure()

plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp, shading='gouraud')

plt.ylim([f[1], f[-1]])

plt.title('Spectrogram')

plt.ylabel('Frequency [Hz]')

plt.xlabel('Time [sec]')

plt.yscale('log')
plt.savefig("inversion_1")
plt.show()
```

```python
#Check non-zero overlap add constraint
signal.check_NOLA(signal.windows.hann(nperseg),nperseg,nperseg // 2)
```

```python
#Zero the components that are 10% or less of the carrier magnitude, then convert back to a time series via inverse STFT
Zxx = np.where(np.abs(Zxx) >= amp/10, Zxx, 0)

_, xrec = signal.istft(Zxx, fs)
```

```python
#Compare the cleaned signal with the original and true carrier signals.
plt.figure()

plt.plot(time, x, time, xrec, time, carrier)

plt.xlim([2, 2.1])

plt.xlabel('Time [sec]')

plt.ylabel('Signal')

plt.legend(['Carrier + Noise', 'Filtered via STFT', 'True Carrier'])

plt.savefig("inversion_2")

plt.show()
```

```python
# Cleaned signal does not start as abruptly as the original, since some of the coefficients of the transient were also removed:
plt.figure()

plt.plot(time, x, time, xrec, time, carrier)

plt.xlim([0, 0.1])

plt.xlabel('Time [sec]')

plt.ylabel('Signal')

plt.legend(['Carrier + Noise', 'Filtered via STFT', 'True Carrier'])
plt.savefig("inversion_3")
plt.show()
```
